2. Resizing to a specific height and width & Preserving the aspect ratio of the image


```python
height = 400
width = 400
dimension = (width, height)
new_image = cv2.resize (img, dimension, interpolation=cv2.INTER_LINEAR)
print("Original shape : ", img.shape)
print("New shape : ", new_image.shape)
cv2_imshow(new_image)
```

3. Resizing with a scalling factor

```python
height = img.shape[0]
width = img.shape[1]

# kita mau ukuran gambarnya jadi 40% dari gambar original

scale_factor = 0.4 #=40%
new_height = int(height * scale_factor) # 300 * 0.4 = 960
new_height = int(width * scale_factor)  # 400 * 0.4 = 1536
dimension = (new_height, new_height)    # (960, 1536)
new_image = cv2.resize (img, dimension, interpolation=cv2.INTER_LINEAR)
print("Original shape : ", img.shape)
print("New shape : ", new_image.shape)
cv2_imshow(new_image)
```

**Mengatur brightness dan contrast dari images**

*   Alpha value untuk controll contrast, range = 1.0-3.0
*   Beta value untuk controll brightness, range = 0-100



```python
def brightness_contrast(img, alpha = 0, beta = 0):
    # alpha contrast control (1.0-3.0)
    # beta brightness control (0-100)

    adjusted = cv2.convertScaleAbs(img, alpha = alpha, beta  = beta)
    return adjusted
```

**Peningkatan contrast**

```python
new_img = brightness_contrast(img, 1, 50) #1 adalah Alpha value dan 50 adalah Beta value

print("Original image")
cv2_imshow(img)

print("\n")

print("Gambar dengan brightness yang ditingkatkan")
cv2_imshow(new_img)
```

```python
new_img = brightness_contrast(img, 2, 0)  #2 adalah Alpha value dan 0 adalah Beta value

print("Original image")
cv2_imshow(img)
cv2.waitKey(0)

print("\n")

print("gambar dengan kontras yang ditingkatkan")
cv2_imshow(new_img)
```

**Membuat gambar menjadi blur**

Gaussian Blur

```python
# menerapkan filter Gaussian dengan ukuran kernel 5x5 & 15x15
blurred1 = cv2.GaussianBlur(img, (5, 5), 0)
blurred2 = cv2.GaussianBlur(img, (15, 15), 0)

# menampilkan gambar asli dan yang telah di-blur
print("Original image")
cv2_imshow(img)

print("\n")

print("Output gaussian blur pertama (5, 5 dan sigma 0)")
cv2_imshow(blurred1)

print("\n")

print("Output gaussian blur kedua (15, 15 dan sigma 0)")
cv2_imshow(blurred2)
```
