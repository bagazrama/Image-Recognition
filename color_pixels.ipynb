```python
#@title Import library { display-mode: "both" }
import PIL 
#  manipulasi gambar
from PIL import Image 
# mengambil gambar dari library
# "PIL"  bisa dimanipulasi
import matplotlib.pyplot as plt 
# untuk melakukan visualisasi gambar yang informatif
```

```python
#@title load image from drive { display-mode: "both" }
# memanggil gambar 
image_path = '/content/drive/MyDrive/Kuliah/S6/PCD/Screenshot (733).png'
image = Image.open (image_path)
pixel_matrix = image.load()
plt.title ('gambar ori')
plt.imshow (image)
plt.show ()
```

```python
display (image)
```

```python
# melihat lebar dan tinggi gambar
width, height = image.size
print(width, height)
```

```python
#@title convertion from RGB to grayscale. R,G,B/3 atau 0.299 * r + 0.587 * g + 0.114 * b { display-mode: "both" }
for i in range(width):
    for j in range(height):
        r, g, b, a = image.getpixel((i, j))
        grayscale = (r + g + b) / 3
        pixel_matrix[i, j] = (int(grayscale), int(grayscale), int(grayscale))

for i in range(width):
    for j in range(height):
        r, g, b, a = image.getpixel((i, j))
        grayscale = 0.299 * r + 0.587 * g + 0.114 * b
        pixel_matrix[i, j] = (int(grayscale), int(grayscale), int(grayscale))

```
